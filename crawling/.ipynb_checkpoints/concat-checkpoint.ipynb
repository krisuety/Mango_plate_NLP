```python
import pandas as pd
df1 = pd.read_csv('mango_review_til_2518.csv')
```

```python
df2 = pd.read_csv('mango_review_til_5000.csv')
```

```python
df2.tail()
```

```python
df = pd.concat([df1, df2])
```

```python
df
```

```python
df['city'] = df['address'].apply(lambda x : x.split(" ")[0])
```

```python
df = df[df['city'] == '서울시']
```

```python
df.drop(columns='city', inplace=True)
```

```python
df['gu'] = df['address'].apply(lambda x : x.split(" ")[1])
```

```python
df = df[df['gu'] != '강남구']
```

```python
df.drop(columns='gu', inplace=True)
```

```python
df
```

```python
df.to_excel("mysql_mango.xlsx", sheet_name='sheet1')
```
